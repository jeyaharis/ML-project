```python
# Complete Machine Learning Model for Fruit Quality Classification
# Enhanced version with nested ZIP file handling for Google Colab

# Install required packages
!pip install -q tensorflow opencv-python matplotlib numpy scikit-learn seaborn

# Import all necessary libraries
import os
import zipfile
import cv2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import shutil

# Function to properly extract nested ZIP structures
def extract_nested_zip(zip_path, extract_to):
    """Extracts zip file and handles nested directory structures"""
    try:
        # First extract the main ZIP
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_to)

        # Check if the extracted content is just one directory
        extracted_items = os.listdir(extract_to)
        if len(extracted_items) == 1:
            # If yes, move its contents up one level
            nested_path = os.path.join(extract_to, extracted_items[0])
            for item in os.listdir(nested_path):
                shutil.move(os.path.join(nested_path, item), os.path.join(extract_to, item))
            os.rmdir(nested_path)

        print(f"Successfully extracted and organized files to {extract_to}")
        return True
    except Exception as e:
        print(f"Error extracting {zip_path}: {str(e)}")
        return False

# Function to load and preprocess images recursively
def load_dataset(folder_path, img_size=(100, 100)):
    """Loads images from directory and preprocesses them (handles nested dirs)"""
    images = []
    labels = []
    class_names = ['good', 'bad']

    for class_idx, class_name in enumerate(class_names):
        class_path = os.path.join(folder_path, class_name)
        if not os.path.exists(class_path):
            print(f"Warning: Missing directory {class_path}")
            continue

        print(f"Loading images from {class_path}...")
        loaded_count = 0

        # Walk through all subdirectories
        for root, _, files in os.walk(class_path):
            for file in files:
                try:
                    # Only process image files
                    if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                        img_path = os.path.join(root, file)
                        img = cv2.imread(img_path)
                        if img is not None:
                            img = cv2.resize(img, img_size)
                            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
                            images.append(img)
                            labels.append(class_idx)
                            loaded_count += 1
                except Exception as e:
                    print(f"Error processing {file}: {str(e)}")

        print(f"Loaded {loaded_count} {class_name} fruit images")

    if len(images) == 0:
        print("\nERROR: No images were loaded. Check your ZIP file structure:")
        print("- Make sure it contains either images directly or one level of subdirectories")
        print("- Supported formats: .jpg, .jpeg, .png")
        return None, None

    # Convert to numpy arrays and normalize
    X = np.array(images, dtype='float32') / 255.0
    y = np.array(labels)

    print(f"\nSuccessfully loaded {len(images)} total images")
    return X, y

# Function to visualize sample images
def show_samples(images, labels, class_names, n_samples=5):
    plt.figure(figsize=(15, 3))
    for i in range(min(n_samples, len(images))):
        plt.subplot(1, n_samples, i+1)
        plt.imshow(images[i])
        plt.title(class_names[labels[i]])
        plt.axis('off')
    plt.show()

# Function to plot training history
def plot_history(history):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train')
    plt.plot(history.history['val_accuracy'], label='Validation')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train')
    plt.plot(history.history['val_loss'], label='Validation')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend()

    plt.show()

# Function to build improved CNN model
def build_model(input_shape, num_classes):
    model = Sequential([
        Conv2D(32, (3, 3), activation='relu', padding='same', input_shape=input_shape),
        MaxPooling2D((2, 2)),
        Conv2D(64, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Conv2D(128, (3, 3), activation='relu', padding='same'),
        MaxPooling2D((2, 2)),
        Flatten(),
        Dense(256, activation='relu'),
        Dropout(0.5),
        Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# Main function
def main():
    # Clean up any previous runs
    !rm -rf good_fruits bad_fruits fruit_dataset

    # Step 1: Upload datasets
    print("=== Dataset Upload ===")
    print("Please upload GOOD fruits dataset (ZIP file):")
    good_zip = files.upload()
    print("\nPlease upload BAD fruits dataset (ZIP file):")
    bad_zip = files.upload()

    # Get the uploaded filenames
    good_zip_name = list(good_zip.keys())[0]
    bad_zip_name = list(bad_zip.keys())[0]

    # Step 2: Extract datasets with nested structure handling
    print("\n=== Extracting Files ===")
    os.makedirs('good_fruits', exist_ok=True)
    os.makedirs('bad_fruits', exist_ok=True)

    extract_nested_zip(good_zip_name, 'good_fruits')
    extract_nested_zip(bad_zip_name, 'bad_fruits')

    # Verify extraction
    print("\nExtracted good fruits files:", len(os.listdir('good_fruits')))
    print("Extracted bad fruits files:", len(os.listdir('bad_fruits')))

    # Step 3: Create organized dataset structure
    print("\n=== Organizing Dataset ===")
    dataset_dir = 'fruit_dataset'
    os.makedirs(dataset_dir, exist_ok=True)
    os.makedirs(os.path.join(dataset_dir, 'good'), exist_ok=True)
    os.makedirs(os.path.join(dataset_dir, 'bad'), exist_ok=True)

    # Move files to appropriate directories (handling both files and directories)
    for item in os.listdir('good_fruits'):
        src = os.path.join('good_fruits', item)
        dst = os.path.join(dataset_dir, 'good', item)
        if os.path.isdir(src):
            shutil.copytree(src, dst, dirs_exist_ok=True)
        else:
            shutil.copy2(src, dst)

    for item in os.listdir('bad_fruits'):
        src = os.path.join('bad_fruits', item)
        dst = os.path.join(dataset_dir, 'bad', item)
        if os.path.isdir(src):
            shutil.copytree(src, dst, dirs_exist_ok=True)
        else:
            shutil.copy2(src, dst)

    # Step 4: Load and preprocess images
    print("\n=== Loading Images ===")
    X, y = load_dataset(dataset_dir, img_size=(128, 128))  # Increased image size

    if X is None or y is None:
        print("\nError: Failed to load dataset. Please check your files.")
        return

    # Step 5: Split dataset
    print("\n=== Preparing Data ===")
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42,
                                                      stratify=y)  # Stratified sampling

    print(f"Training samples: {len(X_train)}")
    print(f"Testing samples: {len(X_test)}")

    # Show sample images
    print("\nSample Images:")
    show_samples(X_train, y_train, ['Good', 'Bad'], n_samples=8)

    # Step 6: Data augmentation
    datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.15,
        height_shift_range=0.15,
        shear_range=0.15,
        zoom_range=0.15,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest'
    )

    # Step 7: Build and train model
    print("\n=== Building Model ===")
    model = build_model(X_train.shape[1:], 2)
    model.summary()

    print("\n=== Training Model ===")
    early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

    history = model.fit(
        datagen.flow(X_train, y_train, batch_size=32),
        epochs=30,
        validation_data=(X_test, y_test),
        callbacks=[early_stopping],
        verbose=1
    )

    # Step 8: Evaluate model
    print("\n=== Model Evaluation ===")
    loss, accuracy = model.evaluate(X_test, y_test)
    print(f"\nTest Accuracy: {accuracy * 100:.2f}%")

    # Plot training history
    plot_history(history)

    # Step 9: Advanced Analysis
    print("\n=== Advanced Analysis ===")

    # Confusion Matrix
    y_pred = np.argmax(model.predict(X_test), axis=1)
    conf_mat = tf.math.confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Good', 'Bad'],
                yticklabels=['Good', 'Bad'])
    plt.title('Confusion Matrix', fontsize=14)
    plt.xlabel('Predicted Label', fontsize=12)
    plt.ylabel('True Label', fontsize=12)
    plt.show()

    # Spoilage Analysis
    spoilage_percent = np.mean(y_pred) * 100
    edible_percent = 100 - spoilage_percent

    # Detailed spoilage report
    print(f"\nDetailed Spoilage Analysis:")
    print(f"- Edible fruits: {edible_percent:.1f}%")
    print(f"- Spoiled fruits: {spoilage_percent:.1f}%")

    # Spoilage visualization
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.pie([edible_percent, spoilage_percent],
            labels=[f'Edible ({edible_percent:.1f}%)',
                   f'Spoiled ({spoilage_percent:.1f}%)'],
            autopct='%1.1f%%',
            colors=['#4CAF50', '#F44336'],
            explode=[0.05, 0],
            shadow=True,
            startangle=90)
    plt.title('Fruit Quality Distribution')

    plt.subplot(1, 2, 2)
    bar_plot = sns.barplot(x=['Good Fruits', 'Spoiled Fruits'],
                          y=[edible_percent, spoilage_percent],
                          palette=['#4CAF50', '#F44336'])
    plt.title('Quality Classification Breakdown')
    plt.ylabel('Percentage (%)')

    for i, v in enumerate([edible_percent, spoilage_percent]):
        bar_plot.text(i, v + 1, f"{v:.1f}%",
                     color='black', ha='center', fontweight='bold')

    plt.tight_layout()
    plt.show()

    print("\n=== Training Complete! ===")

if __name__ == "__main__":
    print("=== Orange Fruit Quality Classification ===")
    main()
```
